## Problème final
**Vous conduisez un peu trop vite et un policier vous arrête. Écrivez une fonction qui retourne l’un des 3 résultats possibles : "No ticket" (Pas d’amende), "Small ticket" (Petite amende) ou "Big Ticket" (Grosse amende).

Si votre vitesse est 60 ou moins, le résultat est "No Ticket".

Si la vitesse est entre 61 et 80 inclus, le résultat est "Small Ticket".

Si la vitesse est 81 ou plus, le résultat est "Big Ticket".

Sauf si c’est votre anniversaire (indiqué par une valeur booléenne dans les paramètres de la fonction) — le jour de votre anniversaire, votre vitesse peut être 5 de plus dans tous les cas.**

```python
def caught_speeding(speed,is_birthday):
    if is_birthday:
        speed -= 5
    if speed <= 60 : return ("No Ticket")
    elif (speed >= 61) and (speed <= 80) : return ("Small Ticket") 
    elif speed >= 81 : return ("Big Ticket")
    pass
```

```python
caught_speeding(81,True)
```

```python
caught_speeding(98,False)
```
